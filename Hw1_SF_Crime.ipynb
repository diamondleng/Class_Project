#### Q3 question (OLAP)
Count the number of crimes each "Sunday" at "SF downtown".   
hint 1: SF downtown is defiend  via the range of spatial location. For example, you can use a rectangle to define the SF downtown, or you can define a cicle with center as well. Thus, you need to write your own UDF function to filter data which are located inside certain spatial range. You can follow the example here: https://changhsinlee.com/pyspark-udf/

hint 2: SF downtown 物理范围可以是 rectangle a < x < b  and c < y < d. thus, San Francisco Latitude and longitude coordinates are: 37.773972, -122.431297. X and Y represents each. So we assume SF downtown spacial range: X (-122.4213,-122.4313), Y(37.7540,37.7740). 也可以是中心一个圈，距离小于多少算做downtown

```
%sql

SELECT Date, DayOfWeek, COUNT(*) AS Count
FROM sf_crime 
WHERE DayOfWeek = 'Sunday'
AND X > -122.4313 
AND X < -122.4213 
AND Y > 37.7540 
AND Y < 37.7740 
GROUP BY Date, DayOfWeek
ORDER BY Count
```

```
%sql

SELECT Date, DayOfWeek, COUNT(*) AS Count
FROM sf_crime 
WHERE DayOfWeek = 'Sunday'
AND X > -122.4313 
AND X < -122.4213 
AND Y > 37.7540 
AND Y < 37.7740 
GROUP BY Date, DayOfWeek
ORDER BY Count
```

The crime amounts in each year did not show a clear trend in either increasing or decreasing. The peak exists in year 2013. The distribution is like a bell distribution.

#### Q7 question (OLAP)
For different category of crime, find the percentage of resolution. Based on the output, give your hints to adjust the policy.

```
%python
Num_Res = spark.sql("Select Category, Resolution, count(*) as N_res from sf_crime group by Category, Resolution")
Num_Cat = spark.sql("select Category, count(*) as N_cat from sf_crime group by Category")
Num_Res.createOrReplaceTempView("Num_Res")
Num_Cat.createOrReplaceTempView("Num_Cat")
```

```
%sql 
SELECT DISTINCT sf_crime.Category, sf_crime.Resolution, N_res/N_cat AS Percentage
FROM (sf_crime LEFT JOIN Num_Res ON sf_crime.Category=Num_Res.Category and sf_crime.Resolution=Num_Res.Resolution) LEFT JOIN Num_Cat on sf_crime.Category=Num_Cat.Category
WHERE sf_crime.Category='LARCENY/THEFT'
ORDER BY Category, Resolution
```

```
%sql 
SELECT DISTINCT sf_crime.Category, sf_crime.Resolution, N_res/N_cat AS Percentage
FROM (sf_crime LEFT JOIN Num_Res ON sf_crime.Category=Num_Res.Category and sf_crime.Resolution=Num_Res.Resolution) LEFT JOIN Num_Cat on sf_crime.Category=Num_Cat.Category
WHERE sf_crime.Category='LARCENY/THEFT'
ORDER BY Category, Resolution
```

We can conclude that 92% of 'LARCENY/THEFT' type of crimes are not resolved. More police effort is suggested on preventing this type of crimes such as adding more police stuffs in most crimes' area or placing more monitors in allowed lots and plazas.
